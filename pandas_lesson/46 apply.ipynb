```python
import pandas as pd 
import numpy as np
import re
```

```python
df = pd.read_table('currency.tsv',
                   index_col='Year')
df
```

```python
df = df.map(lambda v: int(re.sub(r'\D', '', v)))
```

```python
df
```

```python
df.apply(np.sum)
```

```python
df.apply(np.sum, axis = 0 )
```

```python
df.apply(np.sum, axis = 'index' )
```

```python
df.apply(np.sum, axis = 1 )
```

```python
df.apply(np.sum, axis = 'columns' )
```
