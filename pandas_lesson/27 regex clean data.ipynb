```python
url = 'https://en.wikipedia.org/wiki/List_of_unicorn_startup_companies'
dfs=pd.read_html(url,
                match='Valuation', header=0)
len(dfs)
```

```python
df=dfs[0]
df.head()
```

```python
df.columns
```

```python
#ต่อมาเราต้องการที่แก้ไข format ของ Valuation date ให้ถูกต้อง
```

โดยจะต้อง ลบพวก [] ออกไป

```python
df['Valuation date'].str.contains(r'\[')
```

```python
df['Valuation date'] = df['Valuation date'].str.replace(r'\[\d+\]', '')
df['Valuation date'].to_frame()
```

```python
df.drop(columns=['date','aa','test'], inplace=True)
```

```python
df['bb'] = df['Valuation date'].str.contains(r'.*\s\d{4}', case=False)

df
```

```python
df['Valuation date'] = df['Valuation date'].str.extract(r'(.*\s\d{4})', expand=False)
df.drop(columns=['bb'], inplace=True)
df
```

```python
df.columns
```

```python
df[df['Valuation (US$ billions)'].str.match(r'.*\-\d*', case=False)]
df[['valuation min', 'valuation max']] = df['Valuation (US$ billions)'].str.extract(r'(\d+)-(\d+)', expand=True)
df
```

```python
df['valuation min'] = df['Valuation (US$ billions)'].str.extract('(\d+)', expand=False)
df
```

```python
df.columns
```

```python
df.loc[df.isna().any(axis=1),'valuation max'] = df.loc[df.isna().any(axis=1), 'valuation']
df
```
