```python
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(50).cumsum()
ax.plot(data, color='black', linestyle='dashed', label='default')
ax.plot(data, color='blue', linestyle='dashed', drawstyle='steps-post', label='steps-post')
ax.legend()
```
