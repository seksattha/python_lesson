# การใช้ match ร่วมกับ regular expession
\w ก็เป็น text อะไรก็ได้
\w: This represents any word character. Word characters include alphabetic characters (both uppercase and lowercase), digits, and underscores.
$: This asserts the position at the end of the line
[12] matches either "1" or "2"

## Metacharacters:

- **`.` (dot)**: Matches any single character except newline `\n`.
- **`^` (caret)**: Asserts the start of a line.
- **`|` (pipe)**: Acts as an OR operator.
- **`()` (parentheses)**: Groups regular expressions together.
- **`[]` (square brackets)**: Defines a character class.
- **`{}` (curly braces)**: Specifies the number of repetitions or a range of repetitions.
- **`\` (backslash)**: Escapes special characters, or denotes various special sequences.

## Quantifiers:

- **`*`**: Matches zero or more occurrences of the preceding element.
- **`+`**: Matches one or more occurrences of the preceding element.
- **`?`**: Matches zero or one occurrence of the preceding element.
- **`{m}`**: Matches exactly `m` occurrences of the preceding element.
- **`{m, n}`**: Matches at least `m` and at most `n` occurrences of the preceding element.
- **`*?`, `+?`, `??`, `{m,n}?`**: Lazy versions of the above quantifiers, matching as few characters as possible.
