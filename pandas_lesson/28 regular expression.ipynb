## Set of character

```python
df[df['Title'].str.contains(r'[1234you]')]
```

เขียนแบบย่อๆ แบบนี้ก็ได้

```python
df[df['Title'].str.contains(r'[1-4you]')]
```

### and with ?=
`?=` = look ahead  
`.` แทนตัวอักษรอะไรก็ได้  

```python
df[df['Actors'].str.contains(r'.*Downey.*Evans', case=False)]
```

```python
df[df.Actors.str.contains(r'(?=.*Downey)(?=.*Evans)', case=False)]
```

### quantifier
`*` = เป็นตัวขยายตัวอย่างหน้า  
`.` แทนตัวอักษรอะไรก็ได้  
`?` มีอยู่ในนั้นมั้ย มีอยู่ตั้งแต่ 0 ตัว อย่างมาก 1 ตัว
`+` ต้องมีอย่างน้อย 1 ตัวขึ้นไป

`*` = เป็นตัวขยายตัวอย่างหน้า  ในชื่อหนังสือจะมี I กี่ตัวก็ได้ ตั้งแต่ 0 ขึ้นไป

```python
df[df['Title'].str.contains(r'i*', case=False)]
```

เหมือนกับการกำหนดจำหนวนตัวด้วย `{}`

```python
df[df['Title'].str.contains(r'i{2,4}', case=False)]
```

`?` = มีตั้งต่ 0 ตัว อย่างมาก 1 ตัว เขียนเหมืนอกัน i{0,1}

```python
df[df['Title'].str.contains(r'i?', case=False)]
```

`+` = ต้องมีอย่างน้อย 1 ตัว เขียนเหมือนกับ i{1,}

```python
df[df['Title'].str.contains(r'i{2,3}', case=False)]
```

```python
df[df['Title'].str.contains(r'i{2,}', case=False)]
```

```python
df[df['Title'].str.contains(r'\bi{2,}', case=False)]
```

```python
df[df['Title'].str.contains(r'\bi{2,}\b', case=False)]
```
