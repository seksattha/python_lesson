# ALL ANY
## All, any
`any`บอกว่ามี**แถวไหนก็ได้**ที่ได้ตามเงื่อนไง  
`all`**ทุกแถว**จะต้องได้ตามเงื่อนไข ถึงจะบอกว่าเป็น True  
ดูได้ว่ามี**ผลรวม**เท่าไหร่ ด้วยการใช้ `sum `  
(df.sleep_total > 10).sum()

## Checking missing value
ใช้ `ใช้ isna()` จะตรงกับความคิดมากกว่าถ้าเจอว่าเป็นช่องว่างก็ส่งกลับมาเป็น True  
df.isna()  
ถ้าอยากจะดูว่าผลรวมของแต่ละ columns ว่ามีกี่ช่องที่ว่าง 
df.isna().sum()
* ดูว่าแถสไหนมี missing value
 df.isna().any(axis=1)

# ตรวจดูว่าแถวนั้นๆ มี missing value อยู่หรือไม่

```python
df.isna().any(axis=1)
```

## จะแสดงออกมาว่าแถวไหนที่มี ข้อมูลที่มี missing value 

```python
df[df.isna().any(axis=1)]
```

## ในมุมกลับกันก็แสดงได้ว่าแถวไหนมีข้อมูลที่สมบูรณ์แล้วโดยใช้  `notna()`

```python
df.notna().all(axis=1)
```

```python
df[df.notna().all(axis=1)]
```

```python
df.vore.isna().any()
```
