# Reading excel file with pandas

```python
import pandas as pd
```

```python
url ='https://github.com/prasertcbs/basic-dataset/raw/master/Car1_whole2559.xls'
```

```python
url
```

```python
df = pd.read_excel(url)
```

```python
df
```

จะเห็นว่า มันมีจะ row ที่เราไม่ต้องการเข้ามาด้วย 

## การกำหนดว่าจะเลือก sheet ไหน 

#จะต้องมีการใช้ **skiprows** เพื่อที่จะเอา row ที่เราไม่ต้องการออกไป'

```python
df = pd.read_excel(url, sheet_name = 'Total', skiprows = 6)
df
```

## การตัด row ด้านล่างที่ไม่ต้องการ
บางทีในตารางมันจะมีการรวมแล้ว **แล้วเราม่ต้องการ ก็สามารถที่จะเอา row ด้านล่างๆออกไปได้
โดยใช้คำสั่ง skipfooter**

```python
df = pd.read_excel(url,sheet_name = 'Total', skiprows = 6, skipfooter=3)
df
```

## การเลือก column ที่ต้องการ
จะเห็นว่าด้านขวาสุดจะเป็น running number ที่มาจาก excel เราต้องการที่จะตัดมันออกไปอีก 

สามารถทำได้โดยการเลือก column ใน excel ที่เราต้องการมาตั้งแต่แรก **ในทีนี้จะเลือกตั้งแต่ B ถึง H 
โดยใช้ usecols**

```python
df = pd.read_excel(url, sheet_name='Total', usecols='B:I', skiprows=6, skipfooter=3)
df
```

```python
df.columns
```

## การเปลี่ยนชื่อ column name
สร้าง list ของชื่อใหม่ที่ต้องการก่อน
คราวนี้ต้องไม่เอา 7 แถวแรกเพราะว่ามันจะไปเอา header เก่ามาด้วย

```python
columns_name = ['seq','brand','1300cc','1600cc','1800cc','2000cc','gt2000cc','elec','total']
```

```python
df = pd.read_excel(url, sheet_name='Total', usecols='B:I', skiprows=7, skipfooter=3,
                  header = None,
                  names = columns_name)
df
```
