# การลบ index และ column
สามาถที่จะลบได้ทั้ง index และ column **ตั้งค่าได้โดย axis = 0 หรือว่า 1**
โดยที่ 0 จะหมายถึง index อันนี้ไม่จำเป็นจะต้องตั้งค่าอะไรเพราะว่าเครื่องจะ**ตั้งค่าให้ axis =0 โดย default**
เช่นเดียวกันถ้าไม่ได้ตั้ง inplace = True เครื่องจะแค่แสดงออกผลออกมาเฉยๆ ไม่ได้ไปเปลี่ยนค่าใน dataframe แต่อย่างใด

## การลบ Index

#ข้อควรระวังการ ใช้ df.drop เฉยๆ จะไม่ได้เปลี่ยน dataFrame อะไรทั้งสิน.
ถ้าอยากจะให้เปลี่ยนต้องตั้งค่า inplace=True

```python
df.drop(2)
```

```python
df.drop(2, inplace=True)
df
```

## การลบ column

#โดยปกติแล้วมันจะตั้งว่าเป็น axis=0 คือ column **แต่ในกรณีนี้จะต้องตั้งค่าให้เป็น axis=1 column


```python
df.drop('mpg', axis=1,inplace=True)
df
```

```python
df.columns
```

```python
df.drop(['hp' ,'gear' ,'disp'], axis=1, inplace=True)
df
```

```python
df.set_index('model', inplace=True)
df
```
