# Set in to dataframe
sometime it would be difficult ,addressing with number now **using name as index**
##โดยจะใช้คำสั่งอยู่สองอัน
- loc (location)[str] โดยจะใส่เข้าไปเป็น string เช่นในตัวอย่างนี้ก็จะเป็น 'dog' หรือว่า 'Cheetah'
- iloc(integer location) จะระบุเป็นตัวเลข

```python
df.set_index('name', inplace=True)
df
```

ตรงคำสั่ง **inplace=True** หมายถึงให้เปลี่ยนค่าใน Dataframe (update)

## row index addressing

```python
df.loc['Cheetah']
```

**ส่วนใหญ่จะชอบสับสน**
ไปใส่แบบนี้แทน df['dog'] จะได้ error ทันทีเพราะว่าเครื่องจะคิดว่าจะไปดึงข้อมูลใน column มาแทน

```python
df.loc["Cow":'Dog']
```

## Column addressing

```python
df['sleep_total']
```

## addressing both row index and column

```python
df.loc["Cow":'Dog','sleep_total']
```

```python
df.loc["Cow":'Dog',['sleep_total','vore']]
```

**ถ้าจะเอาหลายๆ column จะต้องใช้งานเป็น list**

คราวนี้**เอาทุกแถว**

```python
df.loc[:, 'sleep_total']
```

```python
df.loc[:,'sleep_total':'awake']
```

ตอนนี้ที่ทำเป็น**ช่วงไม่จำเป็นที่จะต้องอยู่ใน list**

```python
%matplotlib inline
df.loc['Cow':'Dog','sleep_total'].plot(kind='barh')
```

```python
%matplotlib inline
df.loc['Cow':'Dog','sleep_total'].nlargest(5).sort_values().plot(kind="barh");
```
