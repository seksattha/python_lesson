# .concat()
joining two dataframe together

```python
import pandas as pd
```

```python
east = pd.read_csv('E.csv')
```

```python
west = pd.read_csv('W.csv')
```

Reading data of provinces from east and west in Thailand and merged into 2 dataframes

joined togother with % `.concat`

There is a reset in index number , to ignore or neglect the inddex just set t

```python
df = pd.concat([west, east])
df
```

```python
df = pd.concat([west, east], ignore_index=True)
df
```

setting the key beforeing joining it together, indexing each dataframe

```python
df = pd.concat([west,east], keys=['ตะวันตก', 'ตะวันออก'])
df
```

clean up some index 

```python
df.reset_index(inplace = True)
```
