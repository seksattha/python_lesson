# Sorting value
การเลือกว่ามากไปน้อยหรือว่าน้อยไปมากสามาที่จะกำหนดได้โดย  
ascending=True/False _โดยถ้าไม่เขียนอะไรเครื่องจะกำหนดให้เป็น true_
- Sort single value
  df.sort_values('sleep_total')  
  df.sort_values('sleep_total', **ascending=False)**
  
  
- sort multiple value
  df.sort_values(['vore','sleep_total'])
  สามารถที่ผ่าน columns เข้าไปหลายๆ column ได้ แต่จะ**ต้องอยู่ใน list**  
  df.sort_values(['vore','sleep_total'], ascending=[True, False])  
  กำหนดมากไปน้อยหรือว่าน้อยไปมาก**แยกกันในแต่ละ column ได้**
- Sort index

  กำหนดถ้าอยากจะ sort index ก็ทำได้เหมืนอเดิม ในกรณีที่เรากำหนด column ที่เราต้องการเป็นหลัก
  df.set_index('name')
  **แค่ใช้ df.sort_index() แค่นี้ไม่ว่าจะสร้างกี่ level**
- Sort column  
  จริงแล้วๆ column มันก็คือ object ที่เป็น index ปรเภทหนึ่ง **แค่ตั้งให้ axis=1**  
  df4 = df3.sort_index **(axis=1)**

  
  

```python
df=pd.read_csv('msleep.csv')
df.head()
```

```python
df.sort_values('sleep_total')
```

```python
df.sort_values('sleep_total', ascending=False)
```

# multiple sort

```python
df.sort_values(['vore','sleep_total'])
```

```python
df.sort_values(['vore','sleep_total'], ascending=[True, False])
```

## Sort by index

```python
df2 = df.set_index('name')
df2
```

```python
df2.sort_index()
```

```python
df2.sort_index(inplace=True)
```

```python
df2
```

```python
df3 = df.set_index(['vore','name'])
df3
```

```python
df3.sort_index()
```

```python
df3.sort_index(inplace=True)
```

```python
df3
```

## columns sorting

```python
df4 = df3.sort_index(axis=1)
df4
```
