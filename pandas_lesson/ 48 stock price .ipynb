```python
import pandas as pd
import pandas_datareader.data as pdr
import datetime as dt
import yfinance as yf
import seaborn as sns
%config InlineBackend.figure.format='retina'
```

```python
stocks = pd.Series(['kbank', 'ptt', 'scc']) + '.bk'
```

```python
stocks
```

```python
try:
    df = yf.download(['kbank.bk','ptt.bk','scc.bk'], start='2015-03-10', end='2024-03-23')
    print(df.head())
except Exception as e:
    print("An error occurred:", e)
```

```python
df
```

```python
ds = df.unstack()['Close'].reset_index().rename(columns={0:'Close', 'Ticker':'Symbols'})
```

```python
ds
```

```python
d_first = ds.groupby('Symbols').first()
```

```python
d_first
```

```python
dx = ds.merge(d_first, on='Symbols', suffixes=('', '_2'))
```

```python
dx
```

```python
dx['chg'] = dx['Close'] - dx['Close_2']
```

```python
dx
```

```python
dx['pct_chg'] = (dx['Close']/dx['Close_2'] - 1) * 100
```

```python
dx
```

```python
sns.relplot(x='Date', y='Close', hue='Symbols', kind='line', data=dx)
```

```python
sns.relplot(x='Date', y='pct_chg', hue='Symbols', kind='line', data=dx)
```

```python
sns.relplot(x='Date', y='pct_chg', col='Symbols', hue='Symbols', kind='line', data=dx)
```
