# การใช้ index slicing 1
`pd.IndexSlice`
df.loc[pd.IndexSlice[:,1977], :]  
[:,1977] *เอาทุกประเทศ ในปี 1997*  
**ไม่จำเป็นที่จะต้องเขียน pd.IndexSlice ทุกครั้ง** สามารถที่จะอ้างเป็นที่เราต้องการได้  
`idx = pd.IndexSlice` กำหนดให้**อ้างเป็น idx**    
**จาก**`df.loc[pd.IndexSlice[:,1977], :]` **มาเป็น**   `df.loc[idx[:,1977:], :]`
* เขียนแบบ`เปรียบเทียบ`
  df.loc[idx[['Thailand','Taiwan'],[1997,2007]],:]




```python
import pandas as pd
```

```python
pd.__version__
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/gapminder.csv',
                index_col = ['country','year'])
df[:50]
```

```python
df.loc['Thailand']
```

```python
df.loc[pd.IndexSlice[:,1977], :]
```

```python
# ไม่ต้องพิมพ์ indexSlice ทุกครั้งก็ได้ อ้างเป็น idx แทน
idx = pd.IndexSlice
```

```python
df.loc[idx[:,1977:], :]
```

```python
df.loc[idx[:,1977],:]
```

```python
df.loc[idx[:,1977:],:]
```

```python
df.loc[idx[:,1977::2],:] # step ทีละ 2
```

```python
df.loc[idx['Taiwan':'Thailand'],:]
```

```python
df.loc[idx['Thailand':,2007],:]
```

```python
df.loc[idx[:,[1977,2007]],:]
```

```python
#เขียนแบบเปรียบเทียบ
df.loc[idx[['Thailand','Taiwan'],[1997,2007]],:]
```

```python
df.loc[idx[['Thailand','Taiwan'],[1997,2007]],'pop']
```

```python
df.loc[idx['T':'U',2007],:] # เอาที่มี 
```

```python
df.loc[idx['T':'U',2007],['pop','lifeExp']]
```
