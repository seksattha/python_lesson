```python
import pandas as pd
```

```python
df = pd.read_csv('IMDB_Movie_1000_Data.csv')
```

```python
df.loc[df.Title.str.contains('man', case=False),['Title', 'Actors']]
```

```python
df.loc[df.Title.str.contains('^man', case=False),['Title', 'Actors']]
```

```python
df.loc[df.Title.str.contains('man$', case=False),['Title', 'Actors']]
```

## Word boundary

```python
df.loc[df['Title'].str.contains(r'man\b', case= False),['Title', 'Actors']]
```

```python
df.loc[df['Title'].str.contains(r'^\bman\b', case= False),['Title', 'Actors']]
```

## lookahead

### positive

ชื่อว่า 'brad' แต่ต้องลงท้ายด้วยคำว่า pitt

```python
df[df['Actors'].str.contains(r'brad\b(?= pitt)', case=False)][['Actors', 'Title']]
```

```python
df.loc[df.Description.str.contains(r'world (?=war)', case=False),['Title','Description']]
```

### negative

ชื่อว่า 'brad' แต่จะต้องไม่ลงท้ายด้วยคำว่า pitt

```python
df[df['Actors'].str.contains(r'brad\b(?! pitt)', case=False)][['Actors', 'Title']]
```

```python
df.loc[df.Director.str.contains(r'Christopher(?! nolan)', case=False),['Director', 'Title']]
```

## lookbehind

### positive

จะต้องลงท้ายด้วย smill โดยที่ตัวก่อนหน้าจะต้องเป็น will

```python
df[df['Actors'].str.contains(r'(?<=will) smith', case=False)][['Actors', 'Title']]
```

### negative

```python
df[df['Actors'].str.contains(r'(?<!Will) Smith', case=False)][['Actors', 'Title']]

```

```python
df.loc[df.Actors.str.contains(r'Diaz', case=False), ['Actors','Title']]
```

```python
df[df['Title'].str.contains(r'(?<!Iron )\bman\b', case=False)][['Actors', 'Title']]
```

## Negates

```python
df[df.Title.str.contains(r'[^\bman\b]')][['Title', 'Actors']]
```
