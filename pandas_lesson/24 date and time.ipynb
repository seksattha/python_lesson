## convert string to date time
จะต้องกำหนด format ให้โดยใช้ string format `strftime`
time string format link [www.google.co.th]  
ถ้าเราไม่กำหนดให้มันจะพยายามเปลี่ยนแปลงให้เราเอง ถ้ามันได้มาตรฐานอยุ่แล้ว 

```python
df.info()
```

```python
df['date_stamp2']=pd.to_datetime(df.date_stamp2, format= '%d-%m-%y')
```

```python
df.info()
```

```python
df.drop(columns = 'date_stame2', inplace=True)
df.info()
```

```python
df['date_stamp3']=pd.to_datetime(df.date_stamp3)
df['iso_date']=pd.to_datetime(df.iso_date)
df['fmt1'] = pd.to_datetime(df.fmt1)
df['fmt2'] = pd.to_datetime(df.fmt2)
df['ym'] = pd.to_datetime(df.ym)
df['ymd'] = pd.to_datetime(df.ymd, format='%Y%m%d')
df['ymdhm'] = pd.to_datetime(df.ymd, format='%Y%m%d%H&M')
df.info()
```

# การใช้ parse date โดย import

```python
df2 = pd.read_table('https://github.com/prasertcbs/tutorial/raw/master/datetime.tsv',
                   index_col = 'date_stamp',
                   parse_dates=True)
df2
```

```python
df3 = pd.read_table('https://github.com/prasertcbs/tutorial/raw/master/datetime.tsv',
                   parse_dates =['date_stamp2', 'date_stamp3', 'iso_date', 'fmt1', 'fmt2', 'ym', 'ymd',
                               'ymdhm'])
df3
```
