```python
import pandas as pd 
```

```python
df = pd.read_csv('mpg.csv')
```

```python
df['trans'] = df.trans.map(lambda v: v[0])
df[['cty', 'hwy']] = df[['cty', 'hwy']].map(lambda v: v * 0.425144 )
```

## the top (n)
รถที่ประหยัดน้ำมันที่สุด 5 อันดับแรก

```python
df.nlargest(5, columns='cty')
```

เพิ่มเงื่อนไข

```python
df.loc[df.trans == 'a',:].nlargest(5, columns='cty')
```

```python
df.loc[df.trans == 'm',:].nlargest(5, columns='cty')
```

แต่ถ้าข้อมูลเยอะๆ แบบนี้คงจะไม่ค่อยสดวกเท่าไหร่ จะต้องใส่การ group เอา

```python
g = df.groupby('trans')
```

```python
g.groups
```

```python
g.get_group('a')
```

```python
g.get_group('m')
```

ถ้าคิดแบบง่ายๆ แบบนี้จะ error ทันที

```python
g.nlargest(3, columns = 'cty')
```

```python
g.apply(lambda grp: grp.nlargest(5, columns = 'cty'))
```

```python
df.groupby('class').apply(lambda grp: grp.nlargest(1, columns = 'cty'))
```
