# Split Apply Combine

```python
import pandas as pd
```

```python
df = pd.read_csv('mpg.csv')
df.sample(5)
```

```python
df.shape
```

```python
df['cty'].mean()
```

การ group เป็นการทำที่เรียกว่า 'split'

```python
g = df.groupby('class')
g.groups
```

```python
g.groups['compact']
```

ถ้าใช้ get group จะได้เป็น dataframe

```python
g.get_group('compact')
```

จะเขียนแบบดิบๆ อย่างนี้เลยก็ได้ 

```python
for k in g.groups.keys():
    print(f'class = {k:3} mean = {g.get_group(k)['cty'].mean()}')
   
```

```python
pandas เขียนอะไรแบบนี้ให้ไว้หมดแล้ว
```

```python
g['cty'].mean()
```

## nlargest()

```python
g=df.groupby('class')[['year', 'model', 'manufacturer', 'drv', 'cty']]
```

```python
g.get_group('compact').nlargest(3,'cty')
```

```python
g.get_group('suv').nlargest(3,'cty')
```

แต่ถ้าอยากได้แบบทีเดียว เขียนแบบนี้ไม่ได้

```python
g.nlargest(3,'cty')
```

```python
g.apply(lambda x: x.nlargest(1,'cty'))
```
