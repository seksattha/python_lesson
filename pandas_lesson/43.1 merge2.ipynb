```python
import pandas as pd
```

```python
data = {'manufacturer':['honda', 'toyoya','mistubishi', 'bmw'],
        'model':['civic','vios','lancer','m2'],
       'price':[1500,1000,748,200]}

data2 = {'manufacturer':['honda', 'toyoya','mistubishi', 'bmw','honda'],
        'mpg':[25,27,34,17,18]}
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
df2 = pd.DataFrame(data2)
df2
```

```python
df_merged = pd.merge(df1, df2, how='outer', on='manufacturer', indicator='true')
df_merged
```

```python
data = {'manufacturer':['honda', 'toyoya','mistubishi', 'bmw'],
        'model':['civic','vios','lancer','m2'],
       'price':[1500,1000,748,200]}

data2 = {'manufacturer':['honda', 'toyoya','mistubishi', 'bmw','honda'],
         'model':['civic','vios','lancer','m2','city'],
        'mpg':[25,27,34,17,18]}
```

ทำการ merge ข้อมูลที่ซ้ำกัน **ไม่ unique** จะทำให้เกิดปัญหาขึ้นทันที

กำหนดให้ df1 เป็นตารางหลัก

```python
df1 = pd.DataFrame(data)
df1
```

สังเกตุได้ว่า ถ้ามีผู้ผลิตชื่อซ้ำกัน แต่ คนละ รุ่น ที่มากกว่าในตารางหลัก

```python
df2 = pd.DataFrame(data2)
df2
```

# ข้อมูลมากกว่า ตารางหลัก

## left join

พอนำมาทำการ merge กันด้วย `left join` ด้วย ผู้ผลิตมันจะเบิ้ลข้อมูลทันที เพราะว่า ข้อมูลมันตรงกัน

```python
df_merged = pd.merge(df1, df2, how='left', on=['manufacturer'], indicator='true')
df_merged
```

## right join

จะได้ช้อมูลเหมือนกัน

```python
df_merged = pd.merge(df1, df2, how='right', on=['manufacturer'], indicator='true')
df_merged
```

## innter join

```python
df_merged = pd.merge(df1, df2, how='inner', on=['manufacturer'], indicator='true')
df_merged
```

## outer join

```python
df_merged = pd.merge(df1, df2, how='outer', on=['manufacturer'], indicator='true')
df_merged
```

**สรุป** ไม่ว่าจะเลือกการ merge แบบไหนก็ไม่สามารถใช้งานได้


# วิธีการแก้ไข 
วิธีการแก้ก็คือจะต้องระบุให้มากกว่าขึ้นทั้ง ผู้ผลิต และ รุ่น  
**`on=[['manufacturer','model']]`**

## left join  
คราวนี้จะได้ข้อมูลที่ถูกต้อง ไม่มีการเบิ้ลแถว

```python
df_merged = pd.merge(df1, df2, how='left', on=['manufacturer','model'], indicator=True)
df_merged
```

## right join  
คราวนี้จะเห็นได้เลยว่า แถวที่เกินออกมา จะไม่มีข้อมูล เพราะว่ามันอยู่ในตารางรองเท่านั้น

```python
df_merged = pd.merge(df1, df2, how='right', on=['manufacturer','model'], indicator=True)
df_merged
```

## inner join  
inner จะได้ผลลัพธ์ที่เหมือนกัน left join

```python
df_merged = pd.merge(df1, df2, how='inner', on=['manufacturer','model'], indicator=True)
df_merged
```

## outer join  
inner จะได้ผลลัพธ์ที่เหมือนกัน right join

```python
df_merged = pd.merge(df1, df2, how='outer', on=['manufacturer','model'], indicator=True)
df_merged
```
