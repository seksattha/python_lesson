การเปลี่ยนชื่อ column 

df.renames(columns={**ตรงนี้ต้องใช้เป็น dictionary**} )

และก็สามาที่จะเปลี่ยนหลายๆ อันพร้อมกันก็ได้
df.renames(columns={from1:to1, from2:to2} )

สามารถที่จะเปลี่ยนอย่างทีเงื่อนไขโดยใช้ regular expression 
ในตัวอย่างจะเป็นการเปลียน ช่องว่างให้เป็น (' ') ให้เป็น underscore ('_')
โดยใช้คำสั่ง **df.columns.str.replace(r' ', '_')**
 `r''` จะหมายถึงการกำหนดให้เป็น raw เขียนยังไงก็อ่านยังงั้น **เป็นมาตรฐานในการเขียน** ใส่ทุกคร้ัง
`r' '` ช่องว่าง (df)
`,` แทนที่
`'_'` ด้วย underscore 


```python
df = pd.read_table('https://github.com/prasertcbs/basic-dataset/raw/master/BaskinRobbins.tsv',
                  nrows=5)
df
```

```python
df.rename(columns={'Flavour':'flavor'}, inplace = True)
df
```

```python
df.columns
```

```python
df.rename(columns={'Calories':'cal','Carbohydrates (g)':'carb'}, inplace=True)
df
```

```python
df.columns.str.replace(r' ', '_')
```

```python
df.columns=df.columns.str.replace(r' ', '_')
df
```

```python
df.columns=df.columns.str.lower()
df
```

```python
df.columns
```
