## loc
- single label
    เรียก "datsun 710 " มาดูข้อมูล

- list of label
- slice object
- boolean array

```python
pd.__version__
```

```python
df = pd.read_csv('mtcars.csv',
                nrows=10,
                index_col='model')
df
```

```python
df.loc['Datsun 710']
```

```python
df.loc[['Datsun 710']]
```

```python
df.loc[['Datsun 710'], :]
```

```python
df.loc[['Datsun 710'], 'mpg']
```

# list of label

```python
df.loc[['Datsun 710','Merc 230']]
```

df.loc[['Datsun 710','Merc 230'],'mpg':'hp']

```python
df.loc[['Datsun 710','Merc 230'],['mpg','hp']]
```

```python
# slice object
```

```python
df.loc['Datsun 710':'Merc 230']
```

```python
df.loc['Datsun 710':'Merc 230','hp':'gear']
```

# boolean array

```python
df.index.str.contains('Merc')
```

```python
df.loc[df.index.str.contains('Merc')]
```

```python
df.loc[df.index.str.contains('Merc'), 'mpg']
```

```python
df.loc[df.index.str.contains('Merc'), 'mpg':'hp']
```
