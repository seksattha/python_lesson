## figure

figure เหมือนเป็นตัวภาพใหญ่ ทีจะควบคุมรูปย่อย แต่ละ axes เหมือนกับเป็น master control 

```python
fig = plt.figure()
fig, ax = plt.subplots()
fig , axes = plt.subplots(2,2)
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

## axes

```python
data = {'a': np.arange(50), 'c':np.random.randint(0, 50, 50), 'd':np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.rand(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry A')
ax.set_ylabel('entry B')
```

```python
x = np.linspace(0,2,100)
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(x, x, label='linear')
ax.plot(x,x**2, label='quadratic')
ax.plot(x,x**3, label='cubic')


# การตกแต่งกราฟ
ax.set_ylabel('y')
ax.set_xlabel('x')
ax.set_title('example plot')
ax.legend()
```

## color 

```python
fig , ax = plt.subplots(figsize=(5,2.7 ))
data1 = np.arange(50)
data2 = np.random.randn(50)

ax.scatter(data1, data2, facecolor='C0', edgecolor='k')
```

## marker
(https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers)

```python
fig , ax = plt.subplots(figsize=(5,2.7 ))
ax.plot(data1, 'o', label='data1')
ax.plot(data2, 'v', label='data2')
```

## axis label and text


```python
mu, sigma = 115,15
x = mu + sigma * np.random.randn(10000)
fig, ax= plt.subplots(figsize=(5, 2.7))
ax.hist(x, bins=50);

ax.set_xlabel('Length[cm]')
ax.set_ylabel('Probability')
ax.set_title('Example')
ax.text(70,.5, r'$\mu=115, \ \ sigma=15$')
```

## Annotation

```python
fig, ax= plt.subplots(figsize=(5, 2.7))
t = np.arange(0,5,0.001)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, linewidth=4)

ax.annotate('local max', xy=(2,1), xytext=(3, 1.75),
           arrowprops=dict(facecolor = 'black', shrink =0.05)) # ใส่ลูกศร
ax.set_ylim(-2,2)
```

## Legend

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(np.random.rand(100), label='data1')
ax.plot(np.random.rand(100),'o', label='data2')
ax.plot(np.random.rand(100),'x', label='data3')
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
line1, = ax.plot(np.random.rand(100), label='data1')
line2, = ax.plot(np.random.rand(100),'o', label='data2')
ax.plot(np.random.rand(100),'x', label='data3')
ax.legend(handles=[line1, line2], labels =['f','u'],title='my legend!', loc='upper right')
```
